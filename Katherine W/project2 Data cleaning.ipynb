## Group by user to get train/test split

```python
groupuser_df = full_merged_df.groupby('User ID')
groupuser_df.head()
```

```python
# Sort the data by 'user_id' and 'timestamp' in ascending order
df = full_merged_df.sort_values(by=['User ID', 'timestamp'])

# Define a function to extract the most recent 20% of interactions for each user
def get_most_recent_20_percent(group):
    # Calculate the number of rows to include (20% of the group size)
    n_recent = max(1, int(len(group) * 0.2))
    # Select the most recent rows
    return group.tail(n_recent)

def get_last_80_percent(group):
    # Calculate the number of rows to include (20% of the group size)
    n_recent = max(int(len(group) * 0.2), int(len(group) ))
    # Select the most recent rows
    return group.tail(n_recent)

# Apply the function to each user group and concatenate the results
most_recent_20_percent_df = df.groupby('User ID', group_keys=False).apply(get_most_recent_20_percent)
last_80_percent_df = df.groupby('User ID', group_keys=False).apply(get_last_80_percent)
last_80_percent_df.head(30)
```

```python
last_80_percent_df.to_csv("user_train_df.csv", index=False)
most_recent_20_percent_df.to_csv("user_test_df.csv", index=False)
```
